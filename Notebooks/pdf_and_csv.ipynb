```python
# read csv files
import csv

# open file with right encoding
f = open("../Resources/example.csv",mode="r",encoding='utf-8')
# get csv reader object
csv_data = csv.reader(f)
# cast it to a list
data_lines = list(csv_data)
data_lines = data_lines[:100]
f.close()

print(data_lines[0:2])

# extract all names and email
mylist = []
for line in data_lines[1:]:
    mylist.append([line[1]+" "+line[2], line[3]])

print(mylist) 

# writing to csv file

f= open("../Resources/output.csv",mode="w")
csv_writer = csv.writer(f,delimiter=",")
csv_writer.writerows(mylist)
f.close()

```

```python
# reading a pdf file
import PyPDF2

f = open("../Resources/Find_the_Phone_Number.pdf", mode= "rb")
pdf_reader = PyPDF2.PdfReader(f)

# get total number of pages
pages = len(pdf_reader.pages)

# get page1 of the file
page1 = pdf_reader.pages[1]

#extract text from page
page1_text = page1.extract_text()

# get all pages text in list
all_pages = pdf_reader.pages
all_text = list(map(lambda x : x.extract_text(),all_pages))
print(all_text)

# EXTRA : finding phone number
pattern = r'\d{3}.\d{3}.\d{4}'
import re
for i in range(len(all_text)):
    matched =re.search(pattern,all_text[i])
    if matched != None:
        phone = matched.group()
        page_to_write = i
        

# write to pdf file
f1= open("../Resources/output.pdf",mode="wb")
pdf_writer = PyPDF2.PdfWriter()

# add pages to new pdf
pdf_writer.add_page(all_pages[page_to_write])
pdf_writer.write(f1)

f.close()
f1.close()

```
